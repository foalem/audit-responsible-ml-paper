```python
def remove_duplicates_and_substrings(csv_file, substrings_to_remove):
    """
    This function reads a CSV file, removes duplicate rows, removes rows where the first column element
    starts with any of a list of substrings, and returns the new data.

    Args:
        csv_file (str): The path to the CSV file to be processed.
        substrings_to_remove (list): A list of substrings to be removed from the first column.

    Returns:
        A pandas DataFrame representing the new CSV file with duplicates and substrings removed.
    """

    # Read in CSV file as a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Drop duplicate rows, keeping the first occurrence of each row
    #df.drop_duplicates(keep='first', inplace=True)

    # Remove rows where the first column element starts with any of the specified substrings
    for substring in substrings_to_remove:
        df = df[~df.iloc[:, 0].str.startswith(substring)]

    return df
```

```python
def add_config_responsable(df1, df2):
    """
    This function finds elements in the 'library' column of the second DataFrame that contain substrings of elements in the
    'config' column of the first DataFrame, and adds two new columns ('config' and 'Responsable_AI_principles') to the second
    DataFrame containing the corresponding 'config' and 'Responsable_AI_principles' values from the first DataFrame.

    Args:
        df1 (pandas.DataFrame): The first DataFrame to use for matching.
        df2 (pandas.DataFrame): The second DataFrame to update.

    Returns:
        The updated second DataFrame with new 'config' and 'Responsable_AI_principles' columns.
    """

    # Initialize new 'config' and 'Responsable_AI_principles' columns in the second DataFrame
    df2['config'] = ''
    df2['Responsable_AI_principles'] = ''

    # Loop over each row in the second DataFrame
    for i, row in df2.iterrows():
        # Loop over each row in the first DataFrame
        for j, row2 in df1.iterrows():
            # Check if the 'config' value in the first DataFrame is a substring of the 'library' value in the second DataFrame
            if row2['config'] in row['library']:
                # If there is a match, set the corresponding 'config' and 'Responsable_AI_principles' values in the second DataFrame
                df2.at[i, 'config'] = row2['config']
                df2.at[i, 'Responsable_AI_principles'] = row2['Responsable_AI_principles']
                # Break out of the inner loop, since we've found a match
                break

    return df2
```

```python
def add_repo_responsable(df1, df2):
    """
    This function finds elements in the 'library' column of the second DataFrame that contain substrings of elements in the
    'config' column of the first DataFrame, and adds two new columns ('config' and 'Responsable_AI_principles') to the second
    DataFrame containing the corresponding 'config' and 'Responsable_AI_principles' values from the first DataFrame.

    Args:
        df1 (pandas.DataFrame): The first DataFrame to use for matching.
        df2 (pandas.DataFrame): The second DataFrame to update.

    Returns:
        The updated second DataFrame with new 'config' and 'Responsable_AI_principles' columns.
    """

    # Initialize new 'config' and 'Responsable_AI_principles' columns in the second DataFrame
    df2['Repos'] = ''

    # Loop over each row in the second DataFrame
    for i, row in df2.iterrows():
        # Loop over each row in the first DataFrame
        for j, row2 in df1.iterrows():
            # Check if the 'config' value in the first DataFrame is a substring of the 'library' value in the second DataFrame
            if row2['Repos'] in row['file']:
                # If there is a match, set the corresponding 'config' and 'Responsable_AI_principles' values in the second DataFrame
                df2.at[i, 'Repos'] = row2['Repos']
                # Break out of the inner loop, since we've found a match
                break

    return df2
```

```python
def plot_repos_by_number_of_principles(df):
    # Get number of unique principles per repo
    repo_principles = df.groupby('Repos')['Responsable_AI_principles'].nunique().reset_index()

    # Get counts of repos with a given number of unique principles
    repo_counts = repo_principles['Responsable_AI_principles'].value_counts().reset_index()
    repo_counts.columns = ['Number of unique principles', 'Number of projects']

    # Calculate percentage of repos with a given number of unique principles
    total_repos = repo_counts['Number of projects'].sum()
    print(repo_counts)
    print("Example")
    repo_counts['Percentage'] = repo_counts['Number of projects'] / total_repos * 100
    print(repo_counts)
    print(total_repos)
    plt.rcParams.update({'font.size': 9})
    fig, ax = plt.subplots(figsize=(5, 3), dpi=500)

    # Plot bar chart
    ax = sns.barplot(x='Number of unique principles', y='Number of projects', data=repo_counts)
    plt.title('Number of project by Number of Responsible AI Principles')

    # Add percentage on top of each bar
    for p in ax.patches:
        height = p.get_height()
        print(height)
        width = p.get_width()
        percentage = f'{p.get_height() / total_repos * 100:.1f}%'
        ax.text(x=p.get_x() + width / 2,
                y=height,
                s=percentage,
                ha='center')

    # Save plot in pdf
    plt.savefig('repos_by_number_of_principles.pdf', dpi=300, bbox_inches='tight')

    plt.show()
```

```python
def draw_and_save_pie_chart(data, filename='pie_chart.pdf'):
    # Convert the data into a pandas DataFrame
    df = pd.DataFrame(list(data.items()), columns=['Category', 'Value'])

    # Set the color palette and increase font size
    sns.set_palette("colorblind")  # Use a vibrant color palette
    plt.rcParams.update({'font.size': 12})  # Increase font size for better readability

    # Create the pie chart
    plt.figure(figsize=(10, 10))
    plt.pie(df['Value'], labels=df['Category'], autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of AI Principles Survey', fontsize=14)

    # Save the plot as a PDF
    plt.savefig(filename, bbox_inches='tight', dpi=300)

    # Show the plot
    plt.show()
```

```python
# Data
data = {
    'Reliability & Safety': 3,
    'Privacy': 2,
    'Fairness': 3,
    'Transparency & Explainability': 1
}

# Draw and save the pie chart
draw_and_save_pie_chart(data, 'ai_principles_survey.pdf')
```
